```python
df = df.drop('Дата публикации объявления', 1)
df = df.drop('Дата обновления объявления', 1)
```

```python
df = df.convert_dtypes()
df.info()
```

```python
df['Рейтинг района'] = df['Рейтинг района'].astype(float)
```

```python
df['Размер комиссии (%)'] = np.where(df['Комиссия'] == 'нет', 0, df['Комиссия'].str.replace('%', ''))
df['Размер комиссии (%)'] = df['Размер комиссии (%)'].astype(int)
df['Комиссия'] = np.where(df['Комиссия'] == 'нет', False, True)
df = df.convert_dtypes()
```

```python
df['Расстояние до ближайшего транспорта'] = df['Расстояние до ближайшего транспорта'].fillna('0')
df['Расстояние до ближайшего транспорта (м)'] = np.where(df['Расстояние до ближайшего транспорта'] is None, '0', 
                                                np.where(df['Расстояние до ближайшего транспорта'].str.endswith('км'), 
                                                df['Расстояние до ближайшего транспорта'].str.replace(' км', '').str.replace('.', '')
                                                         + '00', df['Расстояние до ближайшего транспорта'].str.replace(' м', '')))
df['Расстояние до ближайшего транспорта (м)'] = df['Расстояние до ближайшего транспорта (м)'].astype(int)
df = df.drop('Расстояние до ближайшего транспорта', 1)
df = df.convert_dtypes()
```

```python
df.info()
```

```python
df['Залог (₽)'] = df['Залог (₽)'].fillna(0)
df['Площадь кухни (м²)'] = df['Площадь кухни (м²)'].fillna(0.0)
df['Жилая площадь (м²)'] = df['Жилая площадь (м²)'].fillna(0.0)
df['Материал здания'] = df['Материал здания'].fillna('неизвестно')
df['Ближайший транспорт'] = df['Ближайший транспорт'].fillna('нет')
df.info()
```

```python
df['Количество этажей в доме'] = np.where(df['Этаж'].str.find('/') == 1, df['Этаж'].str[2:], df['Этаж'].str[3:])
df['Этаж'] = np.where(df['Этаж'].str.find('/') == 1, df['Этаж'].str[:1], df['Этаж'].str[:2])
df['Количество этажей в доме'] = df['Количество этажей в доме'].astype(int)
df['Этаж'] = df['Этаж'].astype(int)
df = df.convert_dtypes()
```

```python
df['Тип'] = np.where(df['Комнаты'] == 'Студия', 'Студия', 'Квартира')
df['Комнаты'] = np.where(df['Комнаты'] == 'Студия', 0, df['Комнаты'])
df['Комнаты'] = df['Комнаты'].astype(int)
df = df.convert_dtypes()
```
